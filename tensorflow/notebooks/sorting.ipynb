# Sort Algorithm

```python
## Quick Sort
def swap(a:[], left, right):
    #print(f"swap {a[left]}, {a[right]}")
    m = a[left]
    a[left] = a[right]
    a[right] = m
    
def partition(a: [], pivot: int, left: int, right: int):
    """
    return: index
    a[i>=index]>=pivot
    a[i<index]<pivot
    """
    start = left
    end = right
    swap(a, pivot, end)
    p = a[end]
    while True:
        while left <= end and a[left] < p:
            left += 1
        while right >= start and a[right] >= p:
            right -= 1
        if (left < right):
            swap(a, left, right)
            left += 1
            right -= 1
        else:
            break
    
    index = left + 1
    swap(a, left, end)
    print(f"index: {index}")
    return index

def quicksort(a: [], left: int, right: int):
    if (right > left):
        pivot = (left + right)//2
        print(f"left: {left}, right: {right}， pivot: {pivot}")
        index = partition(a, pivot, left, right)
        print(f"{a}")
        quicksort(a, left, index - 1)
        quicksort(a, index, right)
```
